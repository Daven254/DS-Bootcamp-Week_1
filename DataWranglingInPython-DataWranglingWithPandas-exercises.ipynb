##### Find the number of unique values present in the columns of the DataFrame, store the result in the dictionary format with the name `ex_col_dict` and print the dictionary.
##### Identify and pick the column which has the least number of unique values from the dictionary.
##### Save the column as `ex_grouping_col` and print it.
#### Result:


```python
# iterating through columns to create a dictionary of columns as keys and number of unique values as the value
ex_col_dict = {}
for x in ex_df.columns:
    ex_col_dict[x] = ex_df[x].nunique()

print(ex_col_dict.items())
```

```python
# finding the column with the fewest unique values

# min_val is set to rows in the dataframe. Being that we want the minimum unique values, setting this to the potential maximum allows us to track changes more effectively
min_val = nrows
ex_grouping_col = ''


# iterating through the dictionary and then tracking when we get to the column with the fewest unique value
for key,value in ex_col_dict.items():
    if value < min_val:
        min_val = value
        ex_grouping_col = key
        

        
print(f'We will be grouping based on the {ex_grouping_col} column')
```
